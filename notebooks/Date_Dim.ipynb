```python
#import libraries
from pyspark.sql.functions import sequence, to_date, explode, col, year, month, dayofmonth, dayofweek, weekofyear
```

```python
#generate date


start_year = int(dbutils.widgets.get("start_year"))
end_year = int(dbutils.widgets.get("end_year"))

start_date = f"{start_year}-01-01"
end_date   = f"{end_year}-12-31"

df_date = spark.sql(f"""
    SELECT explode(
        sequence(to_date('{start_date}'), to_date('{end_date}'), interval 1 day)
    ) AS calendar_date
""")
```

```python
#transform

df_date = df_date.select(
    col("calendar_date"),
    year("calendar_date").alias("year"),
    month("calendar_date").alias("month"),
    dayofmonth("calendar_date").alias("day"),
    dayofweek("calendar_date").alias("day_of_week"),
    weekofyear("calendar_date").alias("week_of_year")
)
```

```python
#load
df_date.write.format("delta").mode("overwrite").save('/Volumes/movie_project/gold/date_dim')
```
